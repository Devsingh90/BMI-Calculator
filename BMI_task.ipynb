```python
import json
class BMI: 
    bmi_cat = {
        'UW': 'Under weight',
        'NW': 'Normal weight',
        'OW': 'Over weight',
        'MO': 'Moderately obese',
        'SO': 'Severely obese',
        'VSO': 'Very severely obese'
    }
    hlth_risk = {
        'MR': 'Malnutrition risk',
        'LR': 'Low risk',
        'ER': 'Enhanced risk',
        'MER': 'Medium risk',
        'HR': 'High risk',
        'VHR': 'Very high risk'
    }
# reading Data from json file and appending it to the empty list
def read_data():
    row_list = []
    file_name = 'bmi_data.json'
    with open(file_name,'r', encoding='utf-8') as f:
        data = json.load(f)
#         print('To check input data',data)
        for row in data:
            row_list.append(row)
#         print('reading',row_list)
        return row_list
    

def bmi_data(raw_data):
    start_count = 1
    invalid_data = []
    valid_data = []
    for x in raw_data:
        if start_count == 0:
            start_count = 1
        elif x['Gender'].isalpha() == True and type(x['HeightCm']) == int and type(x['WeightKg']) == int :
            bmi = bmi_calculation(x)
            valid_data.append(x|bmi)
        else:
            invalid_data.append(x)
            start_count +=1
#     To Count Total number of overweight People we can can change the category based on requirement
    count_e = len([i for i in valid_data if i['BMI_categoty'] == 'Over weight'])
    valid_data.append({'Total number of overweight is':count_e})
#     print('To check valid_data',valid_data)
    return valid_data

def bmi_calculation(input_data):
    bmi_category = None
    health_risk = None
    bmi_range = None
    ht = float(input_data['HeightCm'])
    wt = float(input_data['WeightKg'])
    height_mtr_square = ((ht/100)**2)
    bmi = wt/height_mtr_square
    if bmi<= 18.4:
        bmi_category = BMI.bmi_cat.get('UW')
        bmi_range = bmi
        health_risk = BMI.hlth_risk.get('MR')
    elif bmi >= 18.5 and bmi <= 24.9:
        bmi_category = BMI.bmi_cat.get('NW')
        bmi_range = bmi
        health_risk = BMI.hlth_risk.get('LR')
    elif bmi >= 25 and bmi <= 29.9:
        bmi_category = BMI.bmi_cat.get('OW')
        bmi_range = bmi
        health_risk = BMI.hlth_risk.get('ER')
    elif bmi >= 30 and bmi <= 34.9:
        bmi_category = BMI.bmi_cat.get('MO')
        bmi_range = bmi
        health_risk = BMI.hlth_risk.get('MER')
    elif bmi >= 35 and bmi <= 39.9:
        bmi_category = BMI.bmi_cat.get('SO')
        bmi_range = bmi
        health_risk = BMI.hlth_risk.get('HR')
    elif bmi >= 40:
        bmi_category = BMI.bmi_cat.get('VSO')
        bmi_range = bmi
        health_risk = BMI.hlth_risk.get('VHR')
    bmi_dict = {'BMI_categoty':bmi_category,'BMI_range':bmi_range,'Health_risk':health_risk}
#     print("This is Dictionary",bmi_dict)
    return bmi_dict
def write_json(data):
    with open('BMI_final_data.json','w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    return 'Json file added Succesfully..'
if __name__ == '__main__':
    contents = read_data()
#     print(contents)
    BMI_data_final = bmi_data(contents)
#     print('Output in Json Format',BMI_data_final)
    res = write_json(BMI_data_final)
    print(res)
```
